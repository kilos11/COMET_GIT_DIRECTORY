##You can use Comet's table to access and visualize tabular data, logged from a `Pandas dataframe (in the .json, .csv, .md, or .html format)` or from a `file path (with the .csv or .tsv extension)`.

##The following method can be used to log tables:

##*`log_table().`*
##The following panel can be used to visualize tables:

##*`Data Panel.`*
##In addition all the tables logged to an Experiment can be view in the Single Experiment page tab:

##*`Assets & Artifacts Tab: dataframes folder.`*

#**Log a Pandas dataframe**#
##The example below showcases how to log a Pandas dataframe as a table in Comet.

```python
# Import the comet_ml library for experiment tracking
import comet_ml
# Import pandas library for data manipulation
import pandas as pd

# Initialize a new Comet.ml experiment
comet_ml.init()
# Create an experiment object to log data
exp = comet_ml.Experiment()

# Create a sample pandas DataFrame with weekly performance metrics
df = pd.DataFrame(
    {
        "week": ["week 1", "week 2", "week 3"],  # Column for week labels
        "accuracy": [0.85, 0.92, 0.88],  # Column for accuracy values
        "loss": [0.3, 0.2, 0.25]  # Column for loss values
    }
)

# Log the DataFrame as a table in the Comet.ml experiment
exp.log_table(filename="example_table_from_df.csv", tabular_data=df)
```

#**Log a .csv file**#
##The example below showcases how to upload a `.csv` file as a table in Comet.

```python
# Import the comet_ml library for experiment tracking
import comet_ml
# Import pandas library for data manipulation (though not used in this example)
import pandas as pd

# Initialize a new Comet.ml experiment
comet_ml.init()
# Create an experiment object to log data
exp = comet_ml.Experiment()

# Open a new file named "example_table_from_file.csv" in write mode
with open("/tmp/example_table_from_file.csv", "w+") as f:
    # Write the header row to the CSV file
    f.write("week, accuracy, loss\n")
    # Write the first row of data to the CSV file
    f.write("0.85, 0.92, 0.88\n")
    # Write the second row of data to the CSV file
    f.write("0.3, 0.2, 0.25\n")

# Log the CSV file as a table in the Comet.ml experiment
exp.log_table(filename="/tmp/example_table_from_file.csv")
```
