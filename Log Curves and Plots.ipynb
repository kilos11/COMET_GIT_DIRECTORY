#**Log a curve**#
##The example below logs a precision-recall curve from precision and recall metrics calculated with the `sklearn.metrics.precision_recall_curve()` method on 4 mock true and predicted labels.

```python
import comet_ml
import numpy as np
from sklearn.metrics import precision_recall_curve

# Initialize the Comet Experiment
comet_ml.init()
exp = comet_ml.Experiment()

# Create an example x and y vectors for the curve
y_true = np.array([0, 0, 1, 1])
y_predicted = np.array([0.1, 0.4, 0.35, 0.8])
precision, recall, _ = precision_recall_curve(y_true, y_predicted)

# Log the curve to Comet
exp.log_curve(
    name="precision-recall",
    x=precision,
    y= recall,
    step=1,
)
```

**MATPLOTLIB **

```python
 #Import the Comet ML library for experiment tracking
 import matplotlib.pyplot as plt
 # Import the Comet ML library for experiment tracking
 import comet_ml
  # Import NumPy for numerical operations
import numpy as np

# Initialize the Comet Experiment
comet_ml.init()  # Initialize Comet ML
exp = comet_ml.Experiment()  # Create a new Comet experiment

# Create an example scatter plot figure
# Create a dictionary to store our data
data = {
    # Create an array of 50 integers from 0 to 49
    'a': np.arange(50),
    #Create 50 random integers between 0 and 49
    'c': np.random.randint(0, 50, 50),
    # Create 50 random numbers from a normal distribution
    'd': np.random.randn(50)
}

# Create 'b' data with some randomness added to 'a'
data['b'] = data['a'] + 10 * np.random.randn(50)

# Take absolute value of 'd' and scale it up
data['d'] = np.abs(data['d']) * 100

# Create a new figure with 5x5 inch size
fig = plt.figure(figsize=(5, 5))

# Create a scatter plot using our data
plt.scatter('a', 'b', c='c', s='d', data=data)

# Log the created figure to Comet ML
exp.log_figure(figure=fig)
exp.end()  # End the Comet experiment
```

```python
import comet_ml
import plotly.express as px

# Initialize the Comet Experiment
comet_ml.init()
exp = comet_ml.Experiment()

# Create an example figure
df = px.data.gapminder().query("country=='Canada'")
fig = px.line(df, x="year", y="lifeExp", title='Life expectancy in Canada')

# Log the figure to Comet
exp.log_figure(figure=fig, figure_name="plotly_figure.html")
```

SEABORN

```python
import comet_ml
import seaborn as sns

# Initialize the Comet Experiment
comet_ml.init()
exp = comet_ml.Experiment()

# Create an example line plot figure
fmri = sns.load_dataset("fmri")
line_plot = sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event")

# Log the figure to Comet
exp.log_figure(figure=line_plot)
```
