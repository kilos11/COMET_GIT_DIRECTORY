##Comet allows you to create scenes with millions of 3d points that you can visually inspect through advanced functionalities such as zooming, panning, and more. Alternatively to points, you can visualize boxes which are useful when working on segmentation tasks with 3d data.

##The following methods can be used to log 3D point clouds:

##*`log_points_3d()`, either with points or boxes.*
##The following panel can be used to visualize 3D point clouds:

`3d Panel`.
##In addition all 3D point clouds logged to an Experiment can be viewed in the Single Experiment page tab:

`Assets & Artifacts Tab`: others folder.

#**Log with points**#
##The example below logs one 3d cloud defined as a points list using the open3d library.

```python
# Import the comet_ml library for experiment tracking and logging
import comet_ml
# Import numpy for numerical operations
import numpy as np
# Import open3d library for 3D data processing
import open3d as o3d

# Initialize Comet ML for experiment tracking
comet_ml.init()
# Create a new Comet experiment
exp = comet_ml.Experiment()

# Load a sample point cloud dataset from Open3D
dataset = o3d.data.PCDPointCloud()
# Read the point cloud from the dataset file
pcd = o3d.io.read_point_cloud(dataset.path)
# Convert the point cloud to a numpy array and then to a list of points
points = np.asarray(pcd.points).tolist()

# Log the 3D points cloud to Comet ML
exp.log_points_3d(
    # Name of the 3D scene
    scene_name="Open3D Example",
    # List of 3D points
    points=points,
    # Step or iteration number (0 in this case)
    step=0
)
```

#**Log with boxes**#
##The example below logs 100 boxes of equal size, and assigns a random label (and color) to each.

```python
#Import the comet_ml library for experiment tracking and visualization
import comet_ml
# Import the random module for generating random numbers and choices
import random

# Initialize Comet ML for experiment tracking
comet_ml.init()
# Create a new Comet experiment
exp = comet_ml.Experiment()

# Define a dictionary of colors for different object labels
color = {
    "cat": [255, 0, 0],    # Red for cats
    "dog": [0, 255, 0],    # Green for dogs
    "car": [0, 0, 255],    # Blue for cars
    "bicycle": [128, 128, 0],  # Olive for bicycles
                }

# Set the size of each 3D box
box_size = 5
# Initialize an empty list to store the 3D boxes
boxes = []
# Generate 100 random 3D boxes
for i in range(100):
    # Generate a random 3D point (x, y, z) within a 100x100x100 space
    p = [random.randint(0, 100) for v in range(3)]
    # Randomly choose a label from the color dictionary
    label = random.choice(list(color.keys()))

    # Define a single 3D box
    box = {
        # Define the segments (lines) that make up the 3D box
        "segments": [
                # Top face of the box
                [(p[0], p[1], p[2]),  # Starting corner
                (p[0] + box_size, p[1], p[2]),  # Move along x-axis
                (p[0] + box_size, p[1] + box_size, p[2]),  # Move along y-axis
                (p[0], p[1] + box_size, p[2]),  # Move back along x-axis
                (p[0], p[1], p[2]),  # Close the loop
        ],
                # Bottom face of the box
                [
                (p[0], p[1], p[2] + box_size),  # Starting corner (raised by box_size)
                (p[0] + box_size, p[1], p[2] + box_size),  # Move along x-axis
                (p[0] + box_size, p[1] + box_size, p[2] + box_size),  # Move along y-axis
                (p[0], p[1] + box_size, p[2] + box_size),  # Move back along x-axis
                (p[0], p[1], p[2] + box_size),  # Close the loop
                ],
                # Vertical edges connecting top and bottom faces
                [
                [(p[0], p[1], p[2]), (p[0], p[1], p[2] + box_size)],  # Front-left edge
                [(p[0] + box_size, p[1], p[2]), (p[0] + box_size, p[1], p[2] + box_size)],  # Front-right edge
                [(p[0] + box_size, p[1] + box_size, p[2]), (p[0] + box_size, p[1] + box_size, p[2] + box_size)],
                ],
                # Back-right edge
                [
                [(p[0], p[1] + box_size, p[2]), (p[0], p[1] + box_size, p[2] + box_size)],
                ],
                [(p[0], p[1] + box_size, p[2]), (p[0], p[1] + box_size, p[2] + box_size)],  # Back-left edge
                        ],
        "score": random.random(),
        "name": random.choice(["Prediction", "Truth"]),
        "label": label,
        "color": color[label],
    }
    boxes.append(box)

# Log the 3d boxes to Comet
exp.log_points_3d(
    scene_name="3D Boxes Example",
    boxes=boxes,
    step=0,
)
```
