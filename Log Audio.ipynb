#**Log an audio sample**#
##You can log audio from file path or from numpy array. You can find a list of supported audio codecs [link text](https://developer.mozilla.org/en-US/docs/Web/Media/Audio_and_video_delivery/Cross-browser_audio_basics#audio_codec_support).

##The example below logs one randomly generated audio sample as a numpy array.

```python
# Import the comet_ml library for experiment tracking
import comet_ml
# Import numpy for numerical operations
import numpy as np

# Initialize a new Comet.ml experiment
comet_ml.init()
# Create an experiment object to log data
exp = comet_ml.Experiment()

# Set the frequency of the sine wave (in Hz)
fs = 100
# Set the sample rate (number of samples per second)
rate = 44100
# Create a time array from 0 to 1 second with 'rate' number of points
t = np.linspace(0., 1., rate)

# Set the amplitude to the maximum value of 16-bit integer
amplitude = np.iinfo(np.int16).max
# Generate a sine wave audio sample
audio_sample = amplitude * np.sin(2. * np.pi * fs * t)

# Log the generated audio to Comet.ml
exp.log_audio(
    audio_data=audio_sample,  # The audio data to log
    sample_rate=rate,  # The sample rate of the audio
    file_name="Example Audio",  # The name of the audio file
    step=0,  # The step or iteration at which this audio is logged
)
```
